```python
db = pymysql.connect(
    host = '127.0.0.1',
    port = 3306,
    user = 'eddi',
    passwd = 'eddi@123',
    db = 'jpa_test_db'
)

cursor = db.cursor()

sql = """
    create table League(
        id integer not null auto_increment primary key,
        league_title text not null,
        year text not null,
        date text not null,
        time text not null,
        game_state text not null,
        game_title text not null,
        home_team text not null,
        home_team_score text not null,
        away_team_score text not null,
        away_team text not null
    )
"""

cursor.execute(sql)

db.commit()
db.close()

print('League table create success!')
```

```python
f = open('2022_10_league_schedule.csv', 'r', encoding='UTF-8')
csv_reader = csv.reader(f, delimiter=',')

print(csv_reader)
```

```python
cnt = 0

db = pymysql.connect(
    host = '127.0.0.1',
    port = 3306,
    user = 'eddi',
    passwd = 'eddi@123',
    db = 'jpa_test_db'
)

cursor = db.cursor()

for row in csv_reader:
    if cnt == 0:
        cnt += 1
        continue
    
    league_title = (row[1])
    year = (row[2])
    date = (row[3])
    time = (row[4])
    game_state = (row[5])
    game_title = (row[6])
    home_team = (row[7])
    home_team_score = (row[8])
    away_team_score = (row[9])
    away_team = (row[10])
    
#     print(leagueTitle, year, date, time, gameState, gameTitle, homeTeam, homeTeamScore,awayTeamScore, awayTeam)
    
    sql = "INSERT INTO League (league_title, year, date, time, game_state, game_title, home_team, home_team_score, away_team_score, away_team) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
        
    cursor.execute(sql, (league_title, year, date, time, game_state, game_title, home_team, home_team_score, away_team_score, away_team))

db.commit()

f.close()
db.close()
```
