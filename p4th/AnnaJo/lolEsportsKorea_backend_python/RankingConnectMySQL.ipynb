```python
#MySQL table Create
db = pymysql.connect(
    host = '127.0.0.1',
    port = 3306,
    user = 'eddi',
    passwd = 'eddi@123',
    db = 'jpa_test_db'
)

cursor = db.cursor()

sql = """
    create table team_ranking(
        id integer not null auto_increment primary key,
        season_title text not null,
        ranking text not null,
        team_name text not null,
        win_count text not null,
        lose_count text not null,
        goal_difference text not null,
        winning_rate text not null,
        kda_score text not null,
        kill_count text not null,
        death_count text not null,
        assist_count text not null)
"""

cursor.execute(sql)

db.commit()
db.close()

print('TeamRanking table create success!')
```

```python
f = open('2022_LCK_summer_ranking.csv', 'r', encoding='UTF-8')
csv_reader = csv.reader(f, delimiter=',')

print(csv_reader)
```

```python
cnt = 0

db = pymysql.connect(
    host = '127.0.0.1',
    port = 3306,
    user = 'eddi',
    passwd = 'eddi@123',
    db = 'jpa_test_db'
)

cursor = db.cursor()

for row in csv_reader:
    if cnt == 0:
        cnt += 1
        continue
    
    season_title = (row[1])
    ranking = (row[2])
    team_name = (row[3])
    win_count = (row[4])
    lose_count = (row[5])
    goal_difference = (row[6])
    winning_rate = (row[7])
    kda_score = (row[8])
    kill_count = (row[9])
    death_count = (row[10])
    assist_count = (row[11])
        
    sql = "INSERT INTO team_ranking (season_title, ranking, team_name, win_count, lose_count, goal_difference, winning_rate, kda_score, kill_count, death_count, assist_count) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
        
    cursor.execute(sql, (season_title, ranking, team_name, win_count, lose_count, goal_difference, winning_rate, kda_score, kill_count, death_count, assist_count))

db.commit()

f.close()
db.close()
```
